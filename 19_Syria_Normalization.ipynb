データ前処理：Normalization（正規化）

使用技術：NumPy

対象データ：サンプル人口データ

分析手法：Min-Max Normalization

その他：
・人口データをNumPy配列として作成。
・最小値を0、最大値を1とするMin-Max Normalizationを実施。
・異なるスケールの値を比較しやすい形へ変換。
・機械学習やラスタ解析で広く利用される前処理手法。

```python
# Core Libraries
# normalizationに必要なライブラリを読み込む

import numpy as np
```

```python
# ETL Extract
# サンプルの人口データを作る

population = np.array([
    0,
    10,
    50,
    100,
    500
])

print(population)
```

```python
# ETL Transform
# 値を0〜1へ揃える

population_norm = (
    population - population.min()
) / (
    population.max() - population.min()
)

print(population_norm)
```
